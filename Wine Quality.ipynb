**Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
df=pd.read_csv('/content/winequality-red.csv')
```

**Data Visualization**

```python
#Scatter plot
plt.scatter(df['alcohol'],df['pH'],color='g')
plt.xlabel("Alcohol",color='red')
plt.ylabel("Ph value",color='red')
```

```python
#pair plots
import seaborn as sns
sns.pairplot(df)
plt.show()
```

```python
sns.countplot(x=df['quality'],data=df)
```

```python
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
```
